```
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

font = {'family' : 'serif', 'size': 10, 'weight': 'normal'}
plt.rc('font', **font)

fig = plt.figure()
ax = fig.add_subplot()

x = [1, 2, 3, 4, 5, 6]
c = [48.32, 49.87, 51.51, 51.59, 52.98, 49.71]
d = [73.83, 75.16, 77.23, 78.13, 79.01, 78.90]

img = ax.scatter(c, d, c=x, s=50, cmap='cool')
ax.set_xticks([48, 49, 50, 51, 52, 53])
ax.set_xlabel('Exact Match (%)')
ax.set_yticks([73, 74, 75, 76, 77, 78, 79, 80])
ax.set_ylabel('F1-Score (%)')
ax.grid(True, color='#95a5a6', linestyle='--', linewidth=1, alpha=0.5)
handles, labels = img.legend_elements(prop="colors")
legend2 = plt.legend(handles, labels, loc="lower right")

plt.savefig('figure4.pdf')
plt.show()
plt.close()
```
