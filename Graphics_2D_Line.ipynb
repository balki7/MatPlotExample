```
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

font = {'family' : 'serif', 'size': 12, 'weight': 'normal'}
plt.rc('font', **font)

x = [700, 500, 250, 100, 50, 30]
y = [58.25, 60.91, 64.34, 67.21, 75.15, 81.22]

plt.xticks([700, 500, 250, 100, 30])
#plt.yticks([50, 75, 100])
plt.xlabel('Vocabulary File Size (K)')
plt.ylabel('Masked LM Accuracy (%)')
plt.plot(x, y, color='#f266cd', marker='o', linewidth=2, markersize=5)
plt.grid(color='#95a5a6', linestyle='--', linewidth=1, alpha=0.5)

plt.savefig('em_graph.png')
plt.show()
plt.close()
```
