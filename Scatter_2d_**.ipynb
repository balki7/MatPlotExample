```
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

font = {'family' : 'serif', 'size': 10, 'weight': 'normal'}
plt.rc('font', **font)

fig = plt.figure()
ax = fig.add_subplot()

x = [1, 7]
c = [44.89, 48.01]
d = [58.19, 60.65]

img = ax.scatter(c, d, c=x, s=50, cmap='cool')
ax.set_xticks([44, 46, 48, 50, 52, 54, 56])
ax.set_xlabel('Exact Match (%)')
ax.set_yticks([58, 62, 66, 70, 74, 78, 82])
ax.set_ylabel('F1-Score (%)')
ax.grid(True, color='#95a5a6', linestyle='--', linewidth=1, alpha=0.5)
handles, labels = img.legend_elements(prop="colors")
legend2 = plt.legend(handles, labels, loc="lower right")

plt.savefig('figure4.pdf')
plt.show()
plt.close()

x = [2, 4]
c = [57.60, 56.57]
d = [69.13, 67.96]

img = ax.scatter(c, d, c=x, s=50, cmap='cool')
ax.set_xticks([44, 46, 48, 50, 52, 54, 56])
ax.set_xlabel('Exact Match (%)')
ax.set_yticks([58, 62, 66, 70, 74, 78, 82])
ax.set_ylabel('F1-Score (%)')
ax.grid(True, color='#95a5a6', linestyle='--', linewidth=1, alpha=0.5)
handles, labels = img.legend_elements(prop="colors")
legend2 = plt.legend(handles, labels, loc="lower right")

plt.savefig('figure4.pdf')
plt.show()
plt.close()

x = [9]
c = [55.26]
d = [67.07]

img = ax.scatter(c, d, c=x, s=50, cmap='cool')
ax.set_xticks([44, 46, 48, 50, 52, 54, 56])
ax.set_xlabel('Exact Match (%)')
ax.set_yticks([58, 62, 66, 70, 74, 78, 82])
ax.set_ylabel('F1-Score (%)')
ax.grid(True, color='#95a5a6', linestyle='--', linewidth=1, alpha=0.5)
handles, labels = img.legend_elements(prop="colors")
legend2 = plt.legend(handles, labels, loc="lower right")

plt.savefig('figure4.pdf')
plt.show()
plt.close()

x = [3, 5, 6, 8, 10]
c = [52.03, 54.15, 53.69, 53.29, 54.09]
d = [77.71, 78.82, 78.76, 78.48, 79.01]

img = ax.scatter(c, d, c=x, s=50, cmap='cool')
ax.set_xticks([44, 46, 48, 50, 52, 54, 56])
ax.set_xlabel('Exact Match (%)')
ax.set_yticks([58, 62, 66, 70, 74, 78, 82])
ax.set_ylabel('F1-Score (%)')
ax.grid(True, color='#95a5a6', linestyle='--', linewidth=1, alpha=0.5)
handles, labels = img.legend_elements(prop="colors")
legend2 = plt.legend(handles, labels, loc="lower right")

plt.savefig('figure4.pdf')
plt.show()
plt.close()
```
