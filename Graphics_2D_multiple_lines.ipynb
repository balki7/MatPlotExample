```
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

font = {'family' : 'serif', 'size': 12, 'weight': 'normal'}
plt.rc('font', **font)

x = [1, 10, 100, 500, 1000, 3000]
y = [06.03, 13.15, 33.07, 58.96, 67.93, 81.22]
z = [72.5, 73.75, 87.25, 93.52, 95.63, 100]

#plt.xticks([700, 500, 250, 100, 30])
#plt.yticks([50, 75, 100])
plt.xlabel('Number of Training Steps (K)')
plt.ylabel('Accuracy (%)')
plt.plot(x, y, color='#f266cd', marker='o', linewidth=2, markersize=5)
plt.plot(x, z, color='#67b0f5', marker='o', linewidth=2, markersize=5)
plt.grid(color='#95a5a6', linestyle='--', linewidth=1, alpha=0.5)
plt.legend(labels=['MLM', 'NSP'])

plt.savefig('em_graph.png')
plt.show()
plt.close()
```
